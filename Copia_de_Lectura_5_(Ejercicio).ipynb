```python
# dimension del conjunto de datos real,
# mediante el método shape de numpy
# (filas, columnas)

data = dataset.drop(['Hora'], axis=1)
print("La dimension", np.shape(data))

#print(data.columns)

for col in data.columns:
  print(col)
  media = statistics.mean(data[col])
  print("La media es:",media)
  mediana = statistics.median(data[col])
  print("La mediana es:",mediana)
  moda = statistics.mode(data[col])
  print("La moda es:",moda)
```

La 3ra actividad es:
3.- <b>Presente una tabla de distribución de frecuencias para una característica seleccionada.<b>


```python
#Calcular la frecuencia absoluta de una característica
DirectoryEntryExport = data.groupby('Proceso(chrome)\Operaciones de ES de datos/s').agg(frequency=('Proceso(chrome)\Operaciones de ES de datos/s',"count"))
print(DirectoryEntryExport)

DirectoryEntryExport.to_csv("tabla_de_frecuencias_procesoChOperaESDatosS.csv")
```

La 4ta actividad es:
4.- <b>Exhiba el histograma correspondiente a una característica de su elección.<b>


```python
for col in data.columns:

  hist,bin_edges = np.histogram(data[col],10)
  plt.hist(data[col].values)
  plt.title(f"Histograma de la característica {col}")
  plt.grid()
  plt.show()
```

La 5ta actividad es:
5.- <b>Determine los valores de desviación respecto a la media y la desviación media absoluta.<b>


```python

for col in data.columns:
  print(col)
  devest = statistics.stdev(data[col])
  var = statistics.variance(data[col])
  # Calcula la media
  media = statistics.mean(data[col])
  # Calcula la desviación respecto a la media
  desviacion_respecto_media = [x - media for x in data[col]]
  # Calcula la desviación media absoluta
  desviacion_media_absoluta = statistics.mean(map(abs, desviacion_respecto_media))

  print("La media es:",media)
  print("La varianza es:",var)
  print("La desviación estándar es:",devest)
  print("La desviación media absoluta es:", desviacion_media_absoluta)
  print("\n")
```

```python
# dimension del conjunto de datos real,
# mediante el método shape de numpy
# (filas, columnas)

data = dataset.drop(['Hora'], axis=1)
print("La dimension", np.shape(data))

#print(data.columns)

for col in data.columns:
  print(col)
  media = statistics.mean(data[col])
  print("La media es:",media)
  mediana = statistics.median(data[col])
  print("La mediana es:",mediana)
  moda = statistics.mode(data[col])
  print("La moda es:",moda)
```
